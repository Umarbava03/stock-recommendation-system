```python
def second_largest(lis:int):
    lis.sort()
    return lis[-2]
lis=[1,4,2,5]

print(second_largest(lis))
```

```python
def check_addition(lis:int,tar:int):
    for i in range (len(lis)):
        for j in range(i+1):
            if (lis[i]+lis[j])==tar:
                return i,j
    
lis=[1,4,2,5,7,8,9,5]
tar=9
print(check_addition(lis,tar))
```

```python
def dup(lis:int)->int:
    seen=set()

    for i in lis:
        if i in seen:
            return i
        seen.add(i)


lis=[1,4,2,5,7,8,9,5]
print( dup(lis))
```

```python
lis=[1,4,2,5,7,8,9,5]
current_sum=lis[0]
max_sum=lis[0]

for i in lis:
    current_sum=max(current_sum,current_sum+i)
    max_sum=max(max_sum,current_sum)
return max_sum
```
